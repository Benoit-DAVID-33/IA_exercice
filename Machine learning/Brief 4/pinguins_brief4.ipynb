```python
# Basic Libraries
import pandas as pd
import numpy as np

# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import OneHotEncoder

import warnings
warnings.filterwarnings('ignore') 
```

```python
penguins =pd.read_csv("../../CSV/penguins.csv")
penguins
```

```python
penguins = penguins.dropna()
penguins
```

```python
nombre_de_valeurs_uniques = penguins['species'].nunique()

print("Le nombre de valeurs uniques dans la colonne est :", nombre_de_valeurs_uniques)
```

```python
penguins['sex'] = penguins['sex'].factorize()[0]
penguins
```

```python
encoder = OneHotEncoder(sparse=False)

one_hot_encoded = encoder.fit_transform(penguins[['species']])
penguins_encoded = pd.concat([penguins, pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(['species']))], axis=1)

penguins_encoded
```
