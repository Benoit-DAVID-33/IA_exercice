# **Machine Learning supervisé**

La cible (target), appelé aussi y ou output, est connu par l'algorithme d'apprentissage.  
L'algorithme utilise les features (variables), appelé aussi X ou inputs, pour prédire la cible. L'intêret pour l'algorithme, c'est de prédire la cible lorsque celle-ci n'est pas connu, mais lorsque X est connu. Il est alors possible de prédire la cible pour ces observations (échantillons, lignes ou rangées).

# **Machine Learning non supervisé**

Cette fois, il n'y a pas de cible. L'algorithme cherche une relation entre les données. Il est alors possible de déterminer des clusters lorsque les observations sont semblables. Ainsi, On trouve quelles observations sont semblables.

# **Algorithme** vs **Modèle**

Même si on les emploi de manière interchangeable:
* Un algorithme est un "modèle non entraîné".   
* Un algorithme devient un modèle lorsqu'il est entraîné sur des données. Dès lors, il peut faire des prédictions.
