## Overfitting et underfitting


En apprentissage automatique, le **sur-ajustement** (overfitting) se produit lorsqu'un modle d'apprentissage est trop proche des donnes d'apprentissage. Cela conduit  un modle trop complexe.

Le **sous-ajustement** (underfitting) est le contraire. Il se produit lorsqu'un modle d'apprentissage n'est pas assez complexe pour expliquer avec prcision les relations entre les caractristiques d'un ensemble de donnes et une variable cible donne.

**Pour prdire de nouvelles valeurs, nous avons besoin d'un modle bien ajust.**

Pour simplifier, le R2 mesure la diffrence entre les valeurs prdites et les valeurs relles. Un bon score R2 signifie que nous minimisons la distance entre chaque valeur relle et le modle prdictif.

Le R-carr est toujours compris entre 0 et 1 :

## Train Test Split


La solution  notre problme prcdent est le Train Test Split !

1. Tout d'abord, nous divisons nos donnes en deux parties : un ensemble d'entranement (souvent 75% ou 80% de notre ensemble de donnes initial) et un ensemble de test.
2. Ensuite, nous ajustons notre modle uniquement sur l'ensemble d'entrainement. Notre modle ne sait rien des donnes prsentes dans l'ensemble de test.
3. Nous pouvons maintenant calculer le score sur l'ensemble de test. Pour notre modle, il s'agit de nouvelles donnes. Et nous avons des valeurs relles  comparer aux valeurs prdites.
4. Nous comparons notre score d'entrainement avec notre score de test. Si les deux sont  peu prs identiques, il n'y a pas d'overfitting.

Nous devons donc trouver le meilleur mix entre :

- le meilleur score
- la plus petite distance entre le score d'entrainement et le score de test.

Ici, nous n'avons pas confiance dans le modle surajust, car il ne sait pas comment prdire de nouvelles valeurs.
